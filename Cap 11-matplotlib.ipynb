```python
#Gráfico de linha

#Dados
x = np.linspace(0,5,10)
y = x ** 2

#Criar a figura
fig = plt.figure()

#Define a escala dos eixos
axes = fig.add_axes([0,0,0.8,0.8])

#Criar o plot
axes.plot(x,y,'r')

#Labels e título
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de linha');
```

```python
#Gráfico de linha com duas figuras

#Dados
x = np.linspace(0,5,10)
y = x**2

#Criar a figura
fig = plt.figure()

#Criar os eixos
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

#Figura Principal
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes2.set_ylabel('y')
axes1.set_title('Figura Principal')

#Figura Secundária
axes2.plot(y,x,'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Figura Secundária');
```

```python
#Gráfico de linha em paralelo

#Dados
x = np.linspace(0,5,10)
y = x**2

# Divide a área de plotagem em dois subplots
fig, axes = plt.subplots(nrows=1,ncols=2)

# Loop pelos eixos para criar cada plot
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')

#Ajustar o layout
fig.tight_layout()
```

```python
#Gráfico de linha com diferentes escalas

#Dados
x = np.linspace(0,5,10)
y = x**2

# Divide a área de plotagem em dois subplots
fig, axes = plt.subplots(1,2,figsize=(10,4))

#Criar o plot
axes[0].plot(x,x**2,x,np.exp(x))
axes[0].set_title('Escala Padrão')

#Criar o plot 2
axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)');
```

```python
#Grid

#Dados
x = np.linspace(0,5,10)
y = x**2

# Divide a área de plotagem em dois subplots
fig, axes = plt.subplots(1,2,figsize=(10,3))

#Grid padrão
axes[0].plot(x,x**2,x**3,lw=2)
axes[0].grid()

#Grid customizado
axes[1].plot(x,x**2,x**3,lw=2)
axes[1].grid(color='b',alpha=0.7,linestyle='--',linewidth=0.8)
```

```python
#Diferentes estilos de Plots

#Dados
xx= np.linspace(-0.75,1.,100)
n = np.array(range(0,6))

#Subplots
fig, axes = plt.subplots(1,4,figsize=(12,3))

#Plot 1
axes[0].scatter(xx,xx+0.25*np.random.randn(len(xx)),color='k')
axes[0].set_title('scatter')

#Plot 2
axes[1].step(n,n**2,lw=2,color='blue')
axes[1].set_title('step')

#Plot 3
axes[2].bar(n,n**2,align='center',width=0.5,alpha=0.5,color='m')
axes[2].set_title('bar')

#Plot 4
axes[3].fill_between(x,x**2,x**3,alpha=0.5,color='g')
axes[3].set_title('fill_between');
```

```python
#Histogramas

#Dados
n = np.random.randn(100000)

#Subplots
fig, axes = plt.subplots(1,2,figsize=(12,4))

#Plot 1
axes[0].hist(n)
axes[0].set_title('Histograma Padrão')
axes[0].set_xlim((min(n),max(n)))

#Plot 2
axes[1].hist(n,cumulative=True,bins=50)
axes[1].set_title('Histograma Acumulativa')
axes[1].set_xlim((min(n),max(n)));
```

```python
#Dados
alpha = 0.7
phi_ext = 2*np.pi*0.5

# Função para um mapa de cores
def ColorMap(phi_m, phi_p):
    return ( + alpha - 2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

# Mais dados
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T
```

```python
#Criar a figura
fig = plt.figure(figsize=(14,6))

#Adiciona o subplot 1 com projeção 3d
ax = fig.add_subplot(1,2,1,projection='3d')
p = ax.plot_surface(X,Y,Z,rstride=4,cstride=4,linewidth=0)

#Adiciona o subplot 2 com projeção 3d
ax = fig.add_subplot(1,2,2,projection='3d')
p = ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.coolwarm,linewidth=0,antialiased=False)

# Cria a barra de cores como legenda
cb = fig.colorbar(p,shrink=0.5)
```

```python
#wire frame

#Criar a figura
fig = plt.figure(figsize=(8,6))

#Subplot
ax = fig.add_subplot(1,1,1,projection='3d')

#wire frame
p = ax.plot_wireframe(X,Y,Z,rstride=4,cstride=4)
```
