```python
#Carregando um dos datasets que vem com o Seaborn
dados = sns.load_dataset('tips')
```

```python
dados.head()
```

```python
# O mtodo joinplot cria plot de 2 variveis com grficos bivariados e univariados
sns.jointplot(data=dados,x='total_bill',y='tip',kind='reg')
plt.show()
```

```python
# O mtodo lmplot() cria plot com dados e modelos de regresso
sns.lmplot(data=dados,x='total_bill',y='tip',col='smoker')
plt.show()
```

```python
# Construindo um dataframe com Pandas
df = pd.DataFrame()
```

```python
# Alimentando o Dataframe com valores aleatrios
df['Idade'] = rd.sample(range(20,100),30)
df['Massa'] = rd.sample(range(55,150),30)
```

```python
df.shape
```

```python
df.head()
```

```python
# lmplot
sns.lmplot(data=df,x='Idade',y='Massa')
plt.show()
```

```python
#kdeplot
sns.kdeplot(df.Idade)
```

```python
sns.kdeplot(df.Massa)
```

```python
#distplot
sns.distplot(df.Massa)
plt.show()
```

```python
#displot
sns.displot(df.Massa)
plt.show()
```

```python
# Histograma
plt.hist(df.Idade,alpha=0.3)
sns.rugplot(df.Idade)
plt.show()
```

```python
#boxplot
sns.boxplot(df.Idade,color='m')
plt.show()
```

```python
#boxplot
sns.boxplot(df.Massa,color='y')
plt.show()
```

```python
# Violin Plot
sns.violinplot(df.Idade,color='c')
plt.show()
```

```python
# Violin Plot
sns.violinplot(df.Massa,color='g')
```

```python
#Clustermap
sns.clustermap(df)
plt.show()
```

```python
#Clustermap
sns.clustermap(df.T)
plt.show()
```
